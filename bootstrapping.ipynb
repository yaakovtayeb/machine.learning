```python
from random import randint
import numpy as np
n = 1000 # number of examples
# your original set of examples
examples = set(range(n))
results = list()
```

```python
for j in range(1000):
    #  your boostrapped sample
    #give me a random list with k cells
    chosen = [randint(0,n) for k in range(n)]    
    # out-of-sample
    results.append((1000-len(set(chosen)&examples))
                   /float(n))
print ("Out-of-boostrap: %0.1f %%" %
       (np.mean(results)*100))
print(results)
```
